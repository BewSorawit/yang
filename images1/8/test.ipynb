```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'ku.xlsx'

df = pd.read_excel(file_path)

df['date'] = pd.to_datetime(df['date'])

df.set_index('date', inplace=True)

df_resampled = df.resample('7D').sum()  # Use 'sum' instead of 'mean'

# Plot the graph
plt.figure(figsize=(10, 6))

plt.plot(df_resampled.index, df_resampled['tp']*1000, label='Rainfall', marker='o')

plt.title('Rainfall Over Time (7-day Interval)')  # Corrected title
plt.xlabel('Date')
plt.ylabel('Cumulative Rainfall (mm)')  # Adjusted unit to millimeters
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('Cumulative_Rainfall.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  Excel
file_path = 'ku.xlsx'

#  Excel
df = pd.read_excel(file_path)

#  'date'  datetime
df['date'] = pd.to_datetime(df['date'])

#  'date'  index
df.set_index('date', inplace=True)

#  DataFrame  7 
df_resampled = df.resample('7D').agg({'minTemp': 'min', 'maxTemp': 'max', 'meanTemp': 'mean'})

# Plot the graph
plt.figure(figsize=(10, 6))

# Plot minimum temperature
plt.plot(df_resampled.index, df_resampled['minTemp'], label='Min Temperature', marker='o')

# Plot maximum temperature
plt.plot(df_resampled.index, df_resampled['maxTemp'], label='Max Temperature', marker='o')

# Plot mean temperature
plt.plot(df_resampled.index, df_resampled['meanTemp'], label='Mean Temperature', marker='o')

plt.title('Temperature Over Time (7-day Interval)')  
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('Temperature_7day.png', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  Excel
file_path = 'ku.xlsx'

#  Excel
df = pd.read_excel(file_path)

#  'date'  datetime
df['date'] = pd.to_datetime(df['date'])

#  'date'
df.sort_values(by='date', inplace=True)

#  'date'  index
df.set_index('date', inplace=True)

#  DataFrame  7 
df_resampled = df.resample('7D').mean() 

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(df_resampled.index, df_resampled['d2m'], label='Dew Point', marker='o')

#  x  marker 
plt.xticks(df_resampled.index)

# 
plt.title('Weather Factors Over Time (7-day Interval)')
plt.xlabel('Date')
plt.ylabel('Dew Point (C)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

#  PNG
plt.savefig('d2m.png', bbox_inches='tight')

# 
plt.show()
```
