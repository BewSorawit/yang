```python
import pandas as pd
import os
```

```python
target = '../Chanae-regression/season'
```

```python
for file in os.listdir(target):
    df = pd.read_excel(os.path.join(target, file))
    df = df.drop(['date', 'latitude', 'longitude', 'stl1', 'e', 'sp'], axis=1)
    
    # Print the first row of the modified DataFrame
    # print(df.head(1))
    
    # Check columns for numeric types before calculating correlation
    numeric_columns = df.select_dtypes(include='number').columns
    if len(numeric_columns) > 1:  # Ensure there's more than one numeric column
        correlations = df[numeric_columns].corr(method='pearson')
        print(correlations)
    else:
        print("No numeric columns available for correlation.")
    
    print('---------------------------------')
```
